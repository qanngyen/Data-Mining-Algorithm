### Thêm thư viện và đọc dữ liệu

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import fcluster
from sklearn.metrics import silhouette_score
```

### Đọc dữ liệu

```python

file_path = r'D:\Uni\TheThirdYear\kyii\Chinhthuctrentruong\dw\Final_Assignment_9_Group\Algorithms\hierarchical\data\sales_data.csv'  # thay bằng đường dẫn file của bạn
df = pd.read_csv(file_path)
```

### Xây dựng hàm tuyến tính dùng cho việc dự đoán doanh số không bị ảnh hưởng của khuyến mãi

```python
def linear_regression_qty_promo(df):
    results = {}

    # Lấy tất cả các cột QTY và PROMO
    qty_cols = [col for col in df.columns if col.startswith('QTY_')]
    promo_cols = [col for col in df.columns if col.startswith('PROMO_')]

    # Tạo dictionary để dễ lookup PROMO tương ứng
    promo_set = set(promo_cols)

    for qty_col in qty_cols:
        # Tạo tên cột PROMO tương ứng
        promo_col = qty_col.replace('QTY_', 'PROMO_')
        if promo_col in promo_set:
            # Chuẩn bị dữ liệu (bỏ NaN nếu có)
            data = df[[qty_col, promo_col]].dropna()
            if len(data) > 0:
                X = data[[promo_col]].values.reshape(-1,1)
                y = data[qty_col].values

                model = LinearRegression()
                model.fit(X, y)

                a = model.coef_[0]    # Hệ số góc
                b = model.intercept_  # Hệ số chặn

                results[qty_col] = {'coef': a, 'intercept': b}

    return results
results = linear_regression_qty_promo(df)
```

### Tạo ra dữ liệu doanh số không bị ảnh hưởng của khuyến mãi

```python
def adjust_sales_without_promo(df, regression_results, DATE_col='DATE'):
    adjusted_data = {}

    for qty_col, params in regression_results.items():
        promo_col = qty_col.replace('QTY_', 'PROMO_')
        if promo_col in df.columns:
            a = params['coef']
            b = params['intercept']

            adjusted = df[qty_col] - a * df[promo_col]
            adjusted = adjusted.clip(lower=0)
            adjusted = adjusted.apply(np.floor).astype(int)

            adjusted_data[qty_col + '_adjusted'] = adjusted

    # Lấy cột ngày tháng + các cột doanh số điều chỉnh
    df_adjusted_only = pd.DataFrame(adjusted_data)
    df_result = pd.concat([df[[DATE_col]], df_adjusted_only], axis=1)

    return df_result

# Sử dụng hàm
df_adjusted_only = adjust_sales_without_promo(df, results, DATE_col='DATE')

# Lưu file CSV chỉ với doanh số đã điều chỉnh
df_adjusted_only.to_csv(r'D:\Uni\TheThirdYear\kyii\Chinhthuctrentruong\dw\Final_Assignment_9_Group\Algorithms\hierarchical\data\sales_data_without_promo.csv', index=False)
```

### Đọc dữ liệu không bị ảnh hưởng bởi khuyến mãi

```python
# Đọc dữ liệu
data = pd.read_csv(r'D:\Uni\TheThirdYear\kyii\Chinhthuctrentruong\dw\Final_Assignment_9_Group\Algorithms\hierarchical\data\sales_data_without_promo.csv', parse_dates=['DATE'])  # thay bằng tên file thực tế
```

### Chuyển dữ liệu sang dạng dài để có thể thực hiện đọc dữ liệu tốt hơn

```python
# Chuyển dữ liệu từ dạng wide sang long
melted_data = data.melt(id_vars='DATE', var_name='PRODUCT', value_name='QTY')

# Trích xuất tháng
melted_data['MONTH'] = melted_data['DATE'].dt.month
```

### Phân cụm sản phẩm có đặc trưng doanh số giống nhau theo trug bình 1 năm 

```python
# Tính doanh số trung bình mỗi tháng
monthly_avg = melted_data.groupby(['PRODUCT', 'MONTH'])['QTY'].mean().reset_index()

monthly_avg.to_csv(r'D:\Uni\TheThirdYear\kyii\Chinhthuctrentruong\dw\Final_Assignment_9_Group\Algorithms\hierarchical\data\monthly_avg.csv', index=False)


# Pivot lại để tạo ma trận sản phẩm - tháng
product_month_matrix = monthly_avg.pivot(index='PRODUCT', columns='MONTH', values='QTY').fillna(0)
```

```python
# Chuẩn hóa dữ liệu
scaler = StandardScaler()
scaled_matrix = scaler.fit_transform(product_month_matrix)

# Phân cụm bằng Hierarchical Clustering
n_clusters = 5
linkage_matrix = linkage(scaled_matrix, method='ward')
cluster_labels = fcluster(linkage_matrix, n_clusters, criterion='maxclust')

clustered_products = pd.DataFrame({
    'PRODUCT': product_month_matrix.index,
    'CLUSTER': cluster_labels
})

# Lưu ra CSV
clustered_products.to_csv(
    r'D:\Uni\TheThirdYear\kyii\Chinhthuctrentruong\dw\Final_Assignment_9_Group\Algorithms\hierarchical\data\clustered_products.csv',
    index=False,
    encoding='utf-8'
)
print("Đã lưu kết quả phân cụm vào file clustered_products.csv")
```

```python
# Giả sử bạn đã có linkage_matrix từ Hierarchical Clustering
# Chọn số cụm mong muốn (ví dụ: 5)
n_clusters = 5

# Vẽ dendrogram với màu phân cụm rõ ràng
plt.figure(figsize=(14, 6))
dendrogram(
    linkage_matrix,
    labels=product_month_matrix.index.tolist(),
    leaf_rotation=90,
    leaf_font_size=8,
    color_threshold=None,  # Tự động tô màu theo cụm
    above_threshold_color='gray'  # Màu cho nhánh trên ngưỡng
)

# Tự động cắt cụm tại mức tạo ra đúng số cụm mong muốn
# Để vẽ đường cắt ngang:
max_d = linkage_matrix[-(n_clusters-1), 2]  # Khoảng cách tương ứng để tạo n cụm
plt.axhline(y=max_d, c='red', linestyle='--', label=f'Cắt tại khoảng cách = {max_d:.2f}')

plt.title('Dendrogram phân cụm sản phẩm (Hierarchical Clustering)')
plt.xlabel('Sản phẩm')
plt.ylabel('Khoảng cách liên kết')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
score = silhouette_score(scaled_matrix, cluster_labels)
print(f"Silhouette Score = {score:.3f}")
```
